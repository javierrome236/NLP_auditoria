```python
import pandas as pd
import spacy
```

```python
#Cargamos los datos
ds_banco_respuestas = pd.read_csv('./Banco respuestas.txt',sep='\t')
```

```python
#Eliminamos las filas de datos que contengan datos a nulo (Na)
ds_banco_respuestas = ds_banco_respuestas[ds_banco_respuestas['IdPregunta'].notna()]
```

```python
#Mostramos los datos
ds_banco_respuestas
```

```python
#Cargamos el diccionario de palabras del idioma que vayamos a utilizar
nlp = spacy.load('es_core_news_sm')
```

```python
#Tokenizamos las palabras de la primera frase de nuestros datos
doc = nlp(ds_banco_respuestas.iloc[0]['Preguntas'])
```

```python
#Mostramos las diferentes palabras de la frase
for token in doc:
    print(token)
```

```python
#Miramos cuál es el lemma de cada palabra, y si son palabras de relleno o no
for ind in ds_banco_respuestas.index: 
    doc = nlp(ds_banco_respuestas['Preguntas'][ind])
    print(f"Token \t\t\t\tLemma \t\t\t\tStopword".format('Token', 'Lemma', 'Stopword'))
    print("-"*80)
    for token in doc:
        print(f"{str(token)}\t\t\t\t{token.lemma_}\t\t\t\t{token.is_stop}")
```
